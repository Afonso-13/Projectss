
# Pandas Data Visualization Exercise


```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
df = df3.plot.scatter(x='a',y='b',figsize=(12,3),c='red',s=50,edgecolor='black')
df.set_ylim([-0.2,1.2])
df.set_xlim([-0.2,1.2])
```

### Hystogram with pandas

```python
df3['a'].plot.hist(edgecolor='black')
```

```python
with plt.style.context('ggplot'):
    df3['a'].plot.hist(edgecolor='white',alpha=0.55,bins=25)
    
```

### Boxplot

```python
with plt.style.context('ggplot'):
    df3.plot.box(y=['a','b'],showcaps=False,whiskerprops={'linestyle':'--'})
```

### Kde Plot

```python
with plt.style.context('ggplot'):
    df3['d'].plot.kde()
```

```python
with plt.style.context('ggplot'):
    df3['d'].plot.kde(lw=5,ls='--')
```

### AreaPlot

```python
with plt.style.context('ggplot'):
    df=df3.plot.area(alpha=0.4)
    df.set_xlim(0,30)
    df.set_ylim(0,3)
```

```python
with plt.style.context('ggplot'):
    df=df3.plot.area(alpha=0.4)
    df.set_xlim(0,30)
    df.set_ylim(0,3)
    df.legend(loc=(1.02,0.375))
```
